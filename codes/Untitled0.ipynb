```python

import nltk
nltk.download('all')
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords


#df = pd.read_csv("/train.csv")
#first_20_rows = df.head(20)


column_name = "DESCRIPTION"


data =df[column_name].apply(str)


processed_data = []

# Process each row of data
for row in data:
    #individual words
    words = word_tokenize(row)

    # sw = list(sw,wds("eng"))
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.casefold() not in stop_words]

    processed_data.append(" ".join(words))


df['processed_description'] = processed_data
```

```python
from bs4 import BeautifulSoup
import re
import string

def remove_BS(text):
    # Remove HTML Tags 
    soup = BeautifulSoup(text, "html.parser")
    cleaned_text = soup.get_text()

    # Remove Special Characters 
    pattern = r'[^a-zA-Z0-9\s]'
    cleaned_text = re.sub(pattern, '', cleaned_text)
    # return re.sub(pattern, '', cleaned_text)

    # Remove punctations
    text  = "".join([char for char in cleaned_text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    cleaned_text = re.sub('\n' , "", text)

    return cleaned_text

# # Example usage
# html_text = "<p>This is an example <b>HTML</b> text with <a href='https://example.com'>link</a>.</p>"
# me = '<p><strong>!!!!!Aluminum Foil Stickers-good...... ???ASDASD...,,,,as.kitchen helper for mom and wife.&nbsp;</strong><br><br><strong>Specification: &nbsp;</strong><br><strong>'
df1 = df.reset_index()['DESCRIPTION']
df3 = df.reset_index()['BULLET_POINTS']
df2 = df.reset_index()['TITLE']
for i in range(len(df1)):
    cleaned_text = remove_BS(df[i])

# ok = train_df['DESCRIPTION'].apply(remove_BS)
# ok
print(cleaned_text)
```
